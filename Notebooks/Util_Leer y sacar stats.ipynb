```python
import cv2
import os
import matplotlib.pyplot as plt
import pandas as pd
from google.colab.patches import cv2_imshow
import numpy as np
import seaborn as sns
```

```python
rootSourcePath = "/content/drive/Shareddrives/TDG Petroglifos" #Ruta de Unidad compartida
######################################
sourceFolder = "/PetroImages"
######################################
sourcePath = rootSourcePath + sourceFolder
print("Source Path: ", sourcePath)
```

```python
folder_names = os.listdir(sourcePath)
print(folder_names)
```

```python
stats = []
for folder in folder_names:
  folderPath = sourcePath + "/" + folder
  file_names = os.listdir(folderPath)
  for file_name in file_names:
    image_path = sourcePath + "/" + folder + "/" + file_name
    image = cv2.imread(image_path, cv2.IMREAD_UNCHANGED)
    stats.append([file_name, image.shape[0], image.shape[1], image.size])
```

```python
print(stats)
print(len(stats))
```

```python
statsDF = pd.DataFrame(stats)
```

```python
statsDF.columns=["NombreArchivo", "Filas(Alto)", "Columnas(Ancho)", "Size"]
```
