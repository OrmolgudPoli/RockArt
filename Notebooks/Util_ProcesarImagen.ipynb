#0. Organizar en carpetas
Las figuras nuestras estan organizadas en carpetas por grupos.
Se quitan las filas y columnas alrededor de las figuras que no tengan pixeles diferentes a 0 (negro) para después calcular el ancho o alto máximo y agregarles tantas filas y columnas sean necesarias a ambos lados del alto o del ancho para cuadricularlas.

###Importar librerías

```python
import cv2
import os
```

###Definición de rutas
Se define una ruta correspondiente a la carpeta contenedora de las imágenes que se quieren transformar, y una carpeta que es creada donde se guardarán las imágenes transformadas.

```python
#Definición de rutas y creación de directorios
rootSourcePath = "/content/drive/Shareddrives/TDG Petroglifos"
sourceFolder = "/Grupos"

sourcePath = rootSourcePath + sourceFolder
print("Source Path: ",sourcePath)

rootTargetPath = "/content/drive/Shareddrives/TDG Petroglifos"
targetFolder = "/PetroImages"

targetPath = rootTargetPath + targetFolder
print("Target Path: ",targetPath)

if not os.path.exists(targetPath):
    os.makedirs(targetPath)
    print("Directorio creado: ", targetPath)
else:
  print("Directorio ya había sido creado: ", targetPath)
```

###Definición de función

```python
# Función para normalizar imágenes
def preProcesingImages(sourcePath, targetPath):
  # Inicializar un contador para numerar las imágenes
  number = 0
  # Listar todas las carpetas en la ruta de origen y ordenarlas
  folders = os.listdir(sourcePath)
  folders.sort()
  # Iterar a través de cada carpeta
  for folder in folders:
    # Construir la ruta para el grupo actual
    group_path = sourcePath + "/" + folder
    files_names = os.listdir(group_path)
    files_names.sort()
    # Iterar a través de cada archivo en el grupo

    for file_name in files_names:
      img_path = group_path + "/" + file_name
      # Leer la imagen usando OpenCV
      imagecopy = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
      # Extraer solo el primer canal (aplanar la imagen)
      img = imagecopy[:,:,0]

      # Recortar los bordes de la imagen
      sizeOriginal = img.shape[:2]
      heightOriginal = sizeOriginal[0]
      widthOriginal = sizeOriginal[1]
      top = -1
      down = heightOriginal
      row = 0
      left = -1
      right = widthOriginal
      col = 0
      # Encontrar el borde superior
      while top == -1:
        if np.count_nonzero(img[row], axis = 0) != 0:
          top = row
        row += 1
      row = heightOriginal - 1
      # Encontrar el borde inferior
      while down == heightOriginal:
        if np.count_nonzero(img[row], axis = 0) != 0:
          down = row
        row -= 1
      # Encontrar el borde izquierdo
      while left == -1:
        if np.count_nonzero(img[:, col], axis = 0) != 0:
          left = col
        col += 1
      col = widthOriginal - 1
      # Encontrar el borde derecho
      while right == widthOriginal:
        if np.count_nonzero(img[:, col], axis = 0) != 0:
          right = col
        col -= 1
      # Copiar la imagen recortada
      copy = img[top:down + 1, left:right + 1]

      #Squared
      # Redimensionar la imagen para que sea cuadrada
      sizeOriginal = copy.shape[:2]
      heightOriginal = sizeOriginal[0]
      widthOriginal = sizeOriginal[1]
      resized = []
      # Ajustar el ancho o alto según la comparación
      if heightOriginal > widthOriginal: #Rellenar a los lados
        widthHalf = int((heightOriginal - widthOriginal) / 2)
        fill = np.zeros((heightOriginal, widthHalf), int)
        resizedImage = np.hstack((fill, copy))
        if heightOriginal % 2 == 0:
          resizedImage = np.hstack((resizedImage, fill))
        else:
          fill = np.zeros((heightOriginal, widthHalf + 1), int)
          resizedImage = np.hstack((resizedImage, fill))
      elif heightOriginal < widthOriginal: #Rellenar arriba y abajo
        heightHalf = int((widthOriginal - heightOriginal) / 2)
        fill = np.zeros((heightHalf, widthOriginal), int)
        resizedImage = np.vstack((fill, copy))
        if widthOriginal % 2 == 0:
          resizedImage = np.vstack((resizedImage, fill))
        else:
          fill = np.zeros((heightHalf + 1, widthOriginal), int)
          resizedImage = np.vstack((resizedImage, fill))
      else:
        resizedImage = copy
      # Extraer el nombre del archivo sin extensión
      file_name = file_name[:-4]
      if len(str(number)) == 1:
        zeros = "00"
      elif len(str(number)) == 2:
        zeros = "0"
      else:
        zeros = ""
      # Construir la ruta de la carpeta de destino
      targetFolder = targetPath + "/" + zeros + str(number)
      # Crear la carpeta de destino si no existe
      if not os.path.exists(targetFolder):
          os.makedirs(targetFolder)
      # Guardar la imagen procesada en la carpeta de destino
      cv2.imwrite(targetFolder + "/" + folder + file_name + "_" + zeros
                  + str(number) + ".png", resizedImage)
    number = number + 1
```

#Aumentar bordes para rotación

Importar librerias

```python
import cv2
import os
from math import sqrt
import numpy as np
```

Definir rutas

```python
#Definición de rutas y creación de directorios
rootSourcePath = "/content/drive/Shareddrives/TDG Petroglifos"
sourceFolder = "/PetroImages"
sourcePath = rootSourcePath + sourceFolder
print("Source Path: ",sourcePath)

rootTargetPath = "/content/drive/Shareddrives/TDG Petroglifos"
targetFolder = "/PetroImages_v1"
targetPath = rootTargetPath + targetFolder
print("Target Path: ",targetPath)

if not os.path.exists(targetPath):
    os.makedirs(targetPath)
    print("Directorio creado: ", targetPath)
else:
  print("Directorio ya había sido creado: ", targetPath)
```

###Definición de función fillToRotate

```python
#Definición de función fillToRotate
def fillToRotate(sourcePath, targetPath):
  # Obtener la lista de carpetas en la ruta de origen y ordenarlas
  folders = os.listdir(sourcePath)
  folders.sort()
  # Iterar a través de cada carpeta
  for folder in folders:
    # Construir la ruta para el grupo actual
    group_path = sourcePath + "/" + folder
    # Obtener la lista de nombres de archivos en el grupo y ordenarlos
    files_names = os.listdir(group_path)
    files_names.sort()
    # Iterar a través de cada archivo en el grupo
    for file_name in files_names:
      img_path = group_path + "/" + file_name
      imagecopy = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
      # Obtener las dimensiones originales de la imagen
      sizeOriginal = imagecopy.shape[:2]
      heightOriginal = sizeOriginal[0]
      widthOriginal = sizeOriginal[1]
      # Rellenar con una columna o fila en imágenes desfasadas
      if heightOriginal > widthOriginal: #Rellenar a la izquierda
        fill = np.zeros((heightOriginal, 1), int)
        resizedImage = np.hstack((fill, imagecopy))
        side = heightOriginal
      elif heightOriginal < widthOriginal: #Rellenar arriba
        fill = np.zeros((1, widthOriginal), int)
        resizedImage = np.vstack((fill, imagecopy))
        side = widthOriginal
      else:
        resizedImage = imagecopy
        side = widthOriginal
      # Calcular cateto y radio para rellenar
      cathetus = int(side / 2)
      radius = int(sqrt(2 * cathetus ** 2 ))
      # Calcular el ancho lateral para rellenar
      widhtLateral = radius - cathetus
      # Rellenar a un lado de la imagen
      fill = np.zeros((side, widhtLateral), int)
      resizedImage = np.hstack((fill, resizedImage))
      # Calcular el ancho final después de rellenar
      widhtFinal = 2 * widhtLateral + side
      # Agregar bordes
      if widhtFinal % 2 == 0:
        resizedImage = np.hstack((resizedImage, fill))
        fill = np.zeros((widhtLateral, 2 * widhtLateral + side), int)
        resizedImage = np.vstack((fill, resizedImage))
        resizedImage = np.vstack((resizedImage, fill))
      else:
        fill = np.zeros((side, widhtLateral + 1), int)
        resizedImage = np.hstack((resizedImage, fill))
        fill = np.zeros((widhtLateral, 2 * widhtLateral + side + 1), int)
        resizedImage = np.vstack((fill, resizedImage))
        fill = np.zeros((widhtLateral + 1, 2 * widhtLateral + side + 1), int)
        resizedImage = np.vstack((resizedImage, fill))
      # Construir la ruta de la carpeta de destino
      targetFolder = targetPath + "/" + folder
      if not os.path.exists(targetFolder):
          os.makedirs(targetFolder)
      # Eliminar la extensión del nombre del archivo
      file_name = file_name[:-4]
      # Guardar la imagen procesada en la carpeta de destino
      cv2.imwrite(targetFolder + "/" + file_name + ".png", resizedImage)
```

#2. Rectificar tamaño
Las figuras normalmente no quedan centradas, por lo cual se escribe una función para ubicar las columnas y filas en las que está el motivo para eliminar las que sobran.

###Importar librerías

```python
import cv2
import os
import numpy as np
```

###Definición de rutas
Se define una ruta correspondiente a la carpeta contenedora de las imágenes que se quieren transformar, y una carpeta que es creada donde se guardarán las imágenes transformadas.

```python
#Definición de rutas y creación de directorios
rootSourcePath = "/content/drive/Shareddrives/TDG Petroglifos"
sourceFolder = "/MotivosBidimensionales"
sourcePath = rootSourcePath + sourceFolder
print("Source Path: ",sourcePath)

rootTargetPath = "/content/drive/Shareddrives/TDG Petroglifos"
targetFolder = "/MotivosJPGRectificados"
targetPath = rootTargetPath + targetFolder
print("Target Path: ",targetPath)

if not os.path.exists(targetPath):
    os.makedirs(targetPath)
    print("Directorio creado: ", targetPath)
else:
  print("Directorio ya había sido creado: ", targetPath)
```

###Definición de función
La función trimImages procesa imágenes en una carpeta de origen y guarda las versiones recortadas en una carpeta de destino.

```python
#La función trimImages procesa imágenes en una carpeta de origen y guarda las versiones recortadas en una carpeta de destino.
def trimImages(sourcePath, targetPath):
  files_names = os.listdir(sourcePath)
  for file_name in files_names:
    img_path = sourcePath + "/" + file_name
    img = cv2.imread(img_path, cv2.IMREAD_UNCHANGED)
    # Obtener las dimensiones originales de la imagen
    sizeOriginal = img.shape[:2]
    heightOriginal = sizeOriginal[0]
    widthOriginal = sizeOriginal[1]
    # Inicializar variables para determinar las coordenadas de recorte
    top = -1
    down = heightOriginal
    row = 0
    left = -1
    right = widthOriginal
    col = 0
    # Encontrar la coordenada superior del área a recortar
    while top == -1:
      if np.count_nonzero(img[row], axis = 0) != 0:
        top = row
      row += 1
    # Encontrar la coordenada inferior del área a recortar
    row = heightOriginal - 1
    while down == heightOriginal:
      if np.count_nonzero(img[row], axis = 0) != 0:
        down = row
      row -= 1
    # Encontrar la coordenada izquierda del área a recortar
    while left == -1:
      if np.count_nonzero(img[:, col], axis = 0) != 0:
        left = col
      col += 1
    # Encontrar la coordenada derecha del área a recortar
    col = widthOriginal - 1
    while right == widthOriginal:
      if np.count_nonzero(img[:, col], axis = 0) != 0:
        right = col
      col -= 1
    # Copiar el área recortada de la imagen
    copy = img[top:down + 1, left:right + 1]
    # Eliminar la extensión del nombre del archivo
    file_name = file_name[:-4]
    # Guardar la imagen recortada en la carpeta de destino
    cv2.imwrite(targetPath + "/" + file_name + ".png", copy)
```
